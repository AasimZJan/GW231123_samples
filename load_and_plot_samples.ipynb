```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
import corner
import matplotlib
import matplotlib.patches as mpatches
from matplotlib.colors import ListedColormap
from matplotlib import colormaps
from matplotlib.lines import Line2D

def get_handles(colors, labels):
    patches=[]
    for i in np.arange(len(labels)):
        patches.append(mpatches.Patch(color=colors[i], label=labels[i]))
    return patches

def get_samples(data, parameters=['mc', 'eta'], zero_param=None, intrinsic=False):
    final_samples = np.zeros((data.shape[0], len(parameters)))
    rift_parameters = ["m1", "m2", "s1x", "s1y", "s1z", "s2x", "s2y", "s2z", "mc", "eta", "ra", "dec", "time", "phiorb", "incl", "psi", "distance", "Npts", "lnL", "p", "ps", "neff",
    "mtotal", "q", "chi_eff", "chi_p", "m1_source", "m2_source", "mc_source", "mtotal_source", "redshift", "eccentricity", "meanPerAno"]
    if intrinsic:
        rift_parameters = ["m1", "m2", "a1x", "a1y", "a1z", "a2x", "a2y", "a2z", "mc", "eta", "indx", "Npts", "ra", "dec", "tref", "phiorb", "incl", "psi", "dist", "p", "ps", "lnL", "mtotal", "q", "eccentricity", "meanPerAno"]
    for i, param in enumerate(parameters):
        # if a parameter is all zero but you want to plot it, artifically give it a small range so corner doesn't throw an error
        if zero_param is not None:
            if param in zero_param:
                final_samples[:, i] = np.random.uniform(0, 0.000001, final_samples.shape[0])
                continue
        if param == 'chi1':
            final_samples[:, i] = np.sqrt(data[:, 2]**2 + data[:, 3]**2 + data[:, 4]**2)
            continue
        if param == 'chi2':
            final_samples[:, i] = np.sqrt(data[:, 5]**2 + data[:, 6]**2 + data[:, 7]**2)
            continue
        index = rift_parameters.index(param)
        final_samples[:, i] = data[:, index]
    return final_samples
```
